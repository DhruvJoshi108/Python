```python
# The def function in Python is used to define a function. A function is a block of reusable code that performs a specific task.
def even_odd(num):
  if num%2==0:
    print('The Number is Even')
  else:
      print('The Number is Odd')
```

```python
# Call thsi funtion
even_odd(5)
```

```python
def add(d,j):
  return d+j

add(2,3)
```

```python
# Funtion can return multiple values
def add(d,j):
  return d+j,d

add(2,3)
```

```python
# Write a funtion to swap two numbers without using third variable
def swap(a,b):
  a,b=b,a
  return a,b
swap(2,3)
```

```python
#  Write a funtion to swap two numbers with using third variable
def swap(a,b):
  temp = a
  a = b
  b=temp
  return a,b
swap(3,7)
```
