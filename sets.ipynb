```python
# This function adds both of the sets
a = {'Apple' , True , 1 , 0 , 'Radha' , 'Shyam' }
c = {'avi' , 'raju' , 'amam', 'draco'}
print('Shyam' not in a)
a.add('Vrinda')
print(a)
a.update(c)
print(a)
```

```python
# This function adds a set and a list
a = {'Apple' , True , 1 , 0 , 'Radha' , 'Shyam' }
d = ['hello' , 'hi' , 'world' , 'konochiwa']
a.update(d)
print(a)
```

```python
# This function removes a item from the set
a = {'Apple' , True , 1 , 0 , 'Radha' , 'Shyam'}
a.remove(True)
print(a)
```

```python
#This function removes the given item if the given item is not in the set then it will not show the error
a = {'Apple' , True , 1 , 0 , 'Radha' , 'Shyam', }
a.discard('Apple')
print(a)
```

```python
# This function randomly removes an item from the set
a = {'Apple' , True , 1 , 0 , 'Radha' , 'Shyam', }
o = a.pop()
print(o) #removed item
print(a)
```

```python
# This function will delete the set
x = {'Apple' , True , 1 , 0 , 'Radha' , 'Shyam', }
del x
print(x)
```

**Joining Sets**

```python
s = {'Aman' , 'Diya' , 'Sattu' , 'Raju'}
s1 = {1 , 2, 3, 4, 5, 6, 7}
s2 = s.union(s1)
print(s2)
```

```python
s = {'Aman' , 'Diya' , 'Sattu' , 'Raju'} # Joining sets with pipeline method we can join multiple sets togather with this method
s1 = {1 , 2, 3, 4, 5, 6, 7}
s3 = s|s1
print(s3)
```

```python
# This update method can also join  only 2 sets
s = {'Aman' , 'Diya' , 'Sattu' , 'Raju'}
s1 = {1 , 2, 3, 4, 5, 6, 7}

s.update(s1)
print(s)
```

```python
s = {'Aman' , 'Diya' , 'Sattu' , 'Raju' , 7} # & operator
s1 = {1 , 2, 3, 4, 5, 6, 7}
s2 = s & s1
print(s2)
```

```python
s = {'Aman' , 'Diya' , 'Sattu' , 'Raju' , 7} # diffrence method
s1 = {1 , 2, 3, 4, 5, 6, 7}
s3 = s1.difference(s1)
print(s3)
```
